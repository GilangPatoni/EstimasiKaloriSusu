# deskripsi library yang digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# memanggil dataset

```python
df = pd.read_csv('milk comparison 240ml in csv.csv')
```

# visualisasi data

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

# jumlah kalori berdasarkan brand

```python
models = df.groupby('Brand').count()[['Protein (g)']].sort_values(by='Protein (g)',ascending=True).reset_index()
models = models.rename(columns={'Protein (g)':'number Of Brand'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['Brand'], y=models['number Of Brand'], color='royalblue')
plt.xticks(rotation=60)
```

# distribusi jumlah kalori

```python
plt.figure(figsize=(15,5))
sns.distplot(df['Calories'])
```

# seleksi fitur

```python
features = ['Total Fat (g)','Sugar (g)','Total Carbohydrate (g)','Protein (g)']
x = df[features]
y = df['Calories']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linear

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linear

```python
#Total Fat (g) = 8.0, Sugar = 11, Total Carbohydrate (g) = 12, Protein (g) = 8.0                                                                                                                                                                                                                                                                                              
input_data = np.array([[8.0,11,12,8.0]])

prediction = lr.predict(input_data)
print('Estimasi Jumlah Kalori Susu :', prediction)
```
